```python
import bigframes.pandas as pd


# bf.options.bigquery.location = "EU" #this variable is set based on the dataset you chose to query
# bf.options.bigquery.project = "bi-team-189611" #this variable is set based on the dataset you chose to query

query = """
SELECT
extract (month from created_date) as month_event, dashboard_id, dashboard_title, model, user_id, count(slug) as nb_of_jobs
FROM `looker-h-usage-prod-b35aaee2.prod_looker_usage_marts.looker_activity`
WHERE
created_date between "2025-06-01" and '2025-10-30'
group by 1, 2, 3, 4, 5
"""

df = pd.read_gbq(query)   

# df = bf.read_gbq("select * from `looker-h-usage-prod-b35aaee2.prod_looker_usage_marts.looker_studio_activity`")
# BigFrames can work with tables that are too large to fit in the notebook memory.
# Look at the first 20 rows.

# df.head(10)
# bf.__version__
table = df.pivot_table(
          values = ['nb_of_jobs'],
          index = ['dashboard_title'],
          columns = ['month_event'],
          aggfunc = 'sum'
)

print(table)
```
